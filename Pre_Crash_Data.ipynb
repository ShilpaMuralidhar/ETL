# Renaming the columns

# Importing Crash Type Code Definition file

```python
crash_type = pd.read_csv('/Users/shilpa/Desktop/Crash_Type_Code Definitions.csv')
crash_type.head()
```

```python
Merge_Crash = Rename_Pre_Crash.merge(crash_type, left_on='CrashType', right_on='ID')
Merge_Crash = Merge_Crash.drop([ 'ID', 'CrashType'], axis=1)
Merge_Crash = Merge_Crash.rename(columns={"Def": "CrashType"})
Merge_Crash['CrashType'] = [item.split(")", 1)[1] for item in Merge_Crash['CrashType']]
Merge_Crash = Merge_Crash.rename(columns={"Def": "CrashType"})
Merge_Crash
```

# Importing Distraction1 and Distraction 2

```python
distracted_by = pd.read_csv('/Users/shilpa/Library/Containers/com.microsoft.Excel/Data/Desktop/Driver_distracted_by_Code Definitions.csv')
distracted_by.head()
```

```python
Merge_First_Distraction = Merge_Crash.merge(distracted_by, left_on='FirstDistraction', right_on='ID')
Merge_First_Distraction = Merge_First_Distraction.drop([ 'ID', 'FirstDistraction'], axis=1)
Merge_First_Distraction = Merge_First_Distraction.rename(columns={"Def": "FirstDistraction"})
Merge_First_Distraction['FirstDistraction'] = [item.split(")", 1)[1] for item in Merge_First_Distraction['FirstDistraction']]
Merge_First_Distraction = Merge_First_Distraction.rename(columns={"Def": "FirstDistraction"})
Merge_First_Distraction
```

```python
Merge_Second_Distraction = Merge_First_Distraction.merge(distracted_by, left_on='SecondDistraction', right_on='ID')
Merge_Second_Distraction = Merge_Second_Distraction.drop([ 'ID', 'SecondDistraction'], axis=1)
Merge_Second_Distraction = Merge_Second_Distraction.rename(columns={"Def": "SecondDistraction"})
Merge_Second_Distraction['SecondDistraction'] = [item.split(")", 1)[1] for item in Merge_Second_Distraction['SecondDistraction']]
Merge_Second_Distraction = Merge_Second_Distraction.rename(columns={"Def": "SecondDistraction"})
Merge_Second_Distraction.head()
```

```python
DriverAction = pd.read_csv('/Users/shilpa/Desktop/Driver_manuvered_to_avoid_Code Definitions.csv')
DriverAction.head()
```

# Merge Driver Action

```python
Merge_DriverAction = Merge_Second_Distraction.merge(DriverAction, left_on='DriverAction', right_on='ID')
Merge_DriverAction = Merge_DriverAction.drop([ 'ID', 'DriverAction'], axis=1)
Merge_DriverAction = Merge_DriverAction.rename(columns={"Def": "DriverAction"})
Merge_DriverAction['DriverAction'] = [item.split(")", 1)[1] for item in Merge_DriverAction['DriverAction']]
Merge_DriverAction = Merge_DriverAction.rename(columns={"Def": "DriverAction"})
Merge_DriverAction.head()
```
